```python
import CQER
import pandas as pd
import numpy as np
```

```python
# import data
df = pd.read_excel ('data.xlsx')
```

```python
# output (mean scaling before log transformation)
y  = df['TOTEX']
y  = y.to_numpy()
y  = y.tolist()

# inputs
x1  = df['Energy']
x1  = np.asmatrix(x1).T
x2  = df['Length']
x2  = np.asmatrix(x2).T
x3  = df['Customers']
x3  = np.asmatrix(x3).T
x   = np.concatenate((x1, x2, x3), axis=1)
x   = x.tolist()
```

```python
# calculate the model
from pyomo.environ import *
from pyomo.opt import SolverFactory
crt   = "mult"
func  = "cost"
pps   = "crs"
tau   = 0.7
tile  = "expectile"

model = CQER.cqer(y, x, tau, crt, func, pps, tile)

solver_manager = SolverManagerFactory('neos')
results = solver_manager.solve(model, opt='knitro', tee=True)
```

```python
# display estiamtes
model.em.display()
model.ep.display() 
model.b.display() 
```
